```python
import pandas as pd

df = pd.read_csv('cleaned_titanic.csv') #previously edited on lab 4.1_ex.ipynb and lab3.2.py
```

Example 1: Create Family Size Feature

```python
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
df[['SibSp', 'Parch', 'FamilySize']].head()
```

Example 2: Create Age Group Feature

```python
df['AgeGroup'] = pd.cut(df['Age'],bins=[0, 12, 18, 60, 100],labels=['Child', 'Teen', 'Adult', 'Senior'])
df[['Age', 'AgeGroup']].head()
```

Create Family Status Feature (Flag)

Survival-Based Feature

```python
df['IsAlone'] = df['FamilySize'] == 1
df[['FamilySize', 'IsAlone']].head()
```

Extract Title from Name

Titles can reflect social class and gender

```python
df['Title'] = df['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
df[['Name', 'Title']].head()
```

```python
# Save cleaned dataset for future use    
df.to_csv('cleaned_titanic_F.csv', index=False)
```
