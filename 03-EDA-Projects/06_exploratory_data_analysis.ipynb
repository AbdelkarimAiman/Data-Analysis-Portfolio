## Step 1: Import Libraries and Load Dataset

```python
import pandas as pd
titanic = pd.read_csv("titanic_cleaned_F.csv")
```

## Step 2: High-Level Dataset Overview

Generate summary statistics for all numerical variables.

```python
titanic.describe()
```

## Step 3: Understanding Central Tendency

Compare the mean and median of passenger age.

```python
titanic["Age"].mean()
titanic["Age"].median()
```

```python
print(titanic["FamilySize"].mean())
print(titanic["FamilySize"].median())
```

## Step 4: Analyzing Categorical Variables

Examine the distribution of passenger gender and class.

```python
titanic["Sex"].value_counts(normalize=True)
titanic["Pclass"].value_counts(normalize=True)
```

## Step 5: Group-Based Survival Analysis

Task 1: Survival Rate by Gender

```python
titanic.groupby("Sex")["Survived"].mean()
```

Task 2: Survival Rate by Passenger Class

```python
titanic.groupby("Pclass")["Survived"].mean()
```

## Step 6: Comparing Distributions Between Groups

Compare age distributions for survivors vs non-survivors.

```python
titanic.groupby("Survived")["Age"].describe()
```

## Step 7: Fare Analysis by Class

Analyze average ticket fare by passenger class.

```python
titanic.groupby("Pclass")["Fare"].mean()
```

## Step 8: Family Size Analysis

Examine survival rates based on family size.

```python
titanic.groupby("FamilySize")["Survived"].mean()
```
